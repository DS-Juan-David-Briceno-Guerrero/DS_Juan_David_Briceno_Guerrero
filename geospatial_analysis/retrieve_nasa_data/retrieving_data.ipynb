```python
print(len(fil_periods))
print(fil_periods)
```

```python
#get years.
years = []
#Find years..
for i in fil_periods:
    years.append(int(i[0:4]))

#get months.
months = []
#Find months.
for i in fil_periods:
    months.append(int(i[5:7]))
    
#get days.
day = []
for i in fil_periods:
    day.append(int(i[8:len(i)]))
```

```python
#Create dataframe.
df_periods = pd.DataFrame(
    {'periods_for_scraping': fil_periods,
     'years': years,
     'months': months,
     'day': day
    })
df_periods = df_periods.sort_values(['years', 'months','day'],
              ascending = [False, True, True])
```
